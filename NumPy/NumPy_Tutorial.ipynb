# Bluring a image using slice

```python
import matplotlib.pyplot as plt
```

```python
img = plt.imread('blur.jpg')
```

```python
def smooth(img):
    avg_img = (
        img[1:-1,1:-1] #center
        +img[:-2,1:-1] #top
        +img[2:,1:-1] #bottom
        +img[1:-1,:-2] #left
        +img[1:-1,2:] #right
              )/5.0
    return avg_img
```

```python
blurred = smooth(img)
```

```python
plt.imshow(img,cmap=plt.cm.hot)
plt.figure()
plt.imshow(blurred,cmap=plt.cm.hot)
plt.show()
```

```python
img2 = plt.imread('effil.jpeg')
```

```python
blurred2 = smooth(img2)
```

```python
plt.imshow(img2,cmap=plt.cm.hot)
plt.figure()
plt.imshow(blurred2,cmap=plt.cm.hot)
plt.show()
```
